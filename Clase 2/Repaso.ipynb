```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-10,10,100)
y = 0.5 * x
plt.plot(x,y)
plt.ylim(-10,10)
plt.xlim(-10,10)
plt.ylabel('y = x')
plt.xlabel('x')
plt.title(u"Gráfica de una función lineal \n")
plt.show()
```

```python
y = x**3
plt.plot(x, y, 'b--')
plt.ylabel('y = x^3')
plt.xlabel(u'x - 100 números entre -10 y 10')
plt.title(u'Ejemplos de introducción para el curso de Reconocimiento de Patrones')
plt.show()
```

```python
y2 = x
y = np.sin(x)
y1 = np.cos(x)
plt.plot(y, 'ro', y1, 'b-')
plt.ylabel('y = seno(x)')
plt.xlabel(u'x - 100 números entre -10 y 10')
plt.title(u'Ejemplos de introducción para el curso de Reconocimiento de Patrones')
plt.show()
```

```python
mu, sigma = 0, 2

x = np.linspace(-3, 3, 100)
p = ( 1/(np.sqrt(2 * np.pi) * sigma)) * np.exp( (-1/2) * ( ((x - mu) ** 2) / (sigma ** 2) ) )
plt.plot(x, p, 'b--')
plt.show()
```
